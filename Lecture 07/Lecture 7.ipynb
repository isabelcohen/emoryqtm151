(a) Create two normal random variables ...


- Call them "vec_x" and "vec_z"
- To generate each variable, use loc = 2, scale = 5, size = 1000
- Plot a histogram of each variable
    - Label the axes and title
    - Each graph should have a different title
 
To get full points, plot the histograms on a grid (See Lecture 6)

```python
vec_x = np.random.normal(loc = 2, scale = 5, size = 1000)
vec_z = np.random.normal(loc = 2, scale = 5, size = 1000)

list_vars = [vec_x, vec_z]
list_var_names = ["x", "z"]

fig, list_subfig = plt.subplots(1, 2, figsize = (12, 3))
plt.tight_layout()

index = 0
for var in list_vars:
    list_subfig[index].hist(var)
    list_subfig[index].set_xlabel("Value of variable " + list_var_names[index])
    list_subfig[index].set_ylabel("Frequency")
    list_subfig[index].set_title("Distribution of random variable " + list_var_names[index])
    index = index + 1
```

(f) Create a loop that ...

- Reads multiple ".csv" files
- Creates a numeric vector by 
    - Adding/Subtracting/Mutiplying/Diving two or more variables 
- Plots a histogram of the transformed variable
    - Numbers the figures
    - Labels the axes
    - Labels the title

```python
list_csv_files = ["data/features_stored.csv", "data/features.csv"]

index = 0
for files in list_csv_files:
    data = pd.read_csv(files)
    vec_x = (data['weight'])
    vec_y = (data['mpg'])
    vec_z = vec_x + vec_y
    plt.hist(vec_z)
    plt.title("Figure " + str(index + 1))
    plt.xlabel("weight + mpg")
    plt.ylabel("frequency")
    plt.show()
    index = index + 1
    
```
